```python
import matplotlib.pyplot as plt
import numpy as np
import librosa

%matplotlib inline
```

```python
mags = np.load('phase_comparison/mags.npy')
mags = mags.T
mags /= np.max(mags)
mags.shape
```

```python
plt.figure(figsize=(8, 6))
plt.imshow(mags)
```
